```python
# Select features and target variable for regression
reg_cols = ['Total Volume', 'type', 'year', '4046', '4225', '4770', 'region',]
X_reg = df[reg_cols]
y_reg = df['AveragePrice']
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

num_cols = ['AveragePrice', 'Total Volume', '4046', '4225', '4770']
X_reg = df[num_cols]
X_reg = sm.add_constant(X_reg)
vif = pd.DataFrame()
vif["Variable"] = X_reg.columns
vif["VIF"] = [variance_inflation_factor(X_reg.values, i) for i in range(X_reg.shape[1])]

print(vif)
```

```python
#import liabraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

```python
#finding the best random state

MaxAccu = 0
MaxRs = 0

for i in range(1, 200):
    X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(X_reg, y_reg, test_size=0.3, random_state=i)
    lr = LinearRegression()
    lr.fit(X_reg_train, y_reg_train)
    pred = lr.predict(X_reg_test)
    acc = r2_score(y_reg_test, pred)
    if acc > MaxAccu:
        MaxAccu = acc
        MaxRs = i

print("Maximum R2 score is", MaxAccu, "on Random State", MaxRs)
```

```python
# Select features and target variable for classification
classification_cols = ['AveragePrice','Total Volume', 'region', 'year', '4046', '4225', '4770',]
X_clas = df[classification_cols]
y_clas = df['type']
```

```python
# Perform one-hot encoding on categorical columns
X_clas = pd.get_dummies(X_clas)
```

```python
#features scalling 
from sklearn.preprocessing import StandardScaler

nume_columns = ['AveragePrice','Total Volume', 'year', '4046', '4225', '4770']
sc = StandardScaler()
df[nume_columns] = sc.fit_transform(df[nume_columns])
df[nume_columns]
```

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

X_clas = df[nume_columns]
X_clas = sm.add_constant(X_clas)
vif = pd.DataFrame()
vif["Variable"] = X_clas.columns
vif["VIF"] = [variance_inflation_factor(X_clas.values, i) for i in range(X_clas.shape[1])]

print(vif)
```

```python
#multicollinearity is present but cant drop any columns becuase all columns or data is usefull for prediction
```

```python
#value count in target variable
y_clas.value_counts()
```

Modelling

```python
#Oversampling the data
from imblearn.over_sampling import SMOTE

SM = SMOTE()
X_clas, y_clas= SM.fit_resample(X_clas, y_clas)
```

```python
y_clas.value_counts()
```

```python
#import liabraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
#Finding the best randam state

maxAccu = 0
maxRs = 0

for i in range(1, 200):
    X_clas_train, X_clas_test, y_clas_train, y_clas_test = train_test_split(X_clas, y_clas, test_size=0.3, random_state=i)
    RFC = RandomForestClassifier()
    RFC.fit(X_clas_train, y_clas_train)
    pred = RFC.predict(X_clas_test)
    acc = accuracy_score(y_clas_test, pred)
    if acc > maxAccu:
        maxAccu = acc
        maxRs = i

print("Best accuracy is", maxAccu, "at random_state", maxRs)
```

```python
X_clas_train, X_clas_test, y_clas_train, y_clas_test = train_test_split(X_clas,y_clas, test_size=0.3, random_state=maxRs)
```

```python
# Import libraries
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, accuracy_score
```

```python
# Logistic Regression Classifier
LRC = LogisticRegression()
LRC.fit(X_clas_train, y_clas_train)
PredLRC = LRC.predict(X_clas_test)
print(accuracy_score(y_clas_test, PredLRC))
print(confusion_matrix(y_clas_test, PredLRC))
print(classification_report(y_clas_test, PredLRC))
```

```python
#plotting Roc and compare Auc for all the models used
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import LabelEncoder

# Convert labels to binary format
le = LabelEncoder()
y_clas_test_bin = le.fit_transform(y_clas_test)

# Random Forest Classifier
y_clas_pred_prob_RFC = RFC.predict_proba(X_clas_test)[:, 1]
fpr_RFC, tpr_RFC, thresholds_RFC = roc_curve(y_clas_test_bin, y_clas_pred_prob_RFC)
roc_auc_RFC = roc_auc_score(y_clas_test_bin, y_clas_pred_prob_RFC)
plt.plot(fpr_RFC, tpr_RFC, label='Random Forest (AUC = %0.2f)' % roc_auc_RFC)

# Logistic Regression Classifier
y_clas_pred_prob_LRC = LRC.predict_proba(X_clas_test)[:, 1]
fpr_LRC, tpr_LRC, thresholds_LRC = roc_curve(y_clas_test_bin, y_clas_pred_prob_LRC)
roc_auc_LRC = roc_auc_score(y_clas_test_bin, y_clas_pred_prob_LRC)
plt.plot(fpr_LRC, tpr_LRC, label='Logistic Regression (AUC = %0.2f)' % roc_auc_LRC)

# SVM Classifier
y_clas_pred_prob_SVM = SVM.decision_function(X_clas_test)
fpr_SVM, tpr_SVM, thresholds_SVM = roc_curve(y_clas_test_bin, y_clas_pred_prob_SVM)
roc_auc_SVM = roc_auc_score(y_clas_test_bin, y_clas_pred_prob_SVM)
plt.plot(fpr_SVM, tpr_SVM, label='SVM (AUC = %0.2f)' % roc_auc_SVM)

# Gradient Boosting Classifier
y_clas_pred_prob_GB = GB.predict_proba(X_clas_test)[:, 1]
fpr_GB, tpr_GB, thresholds_GB = roc_curve(y_clas_test_bin, y_clas_pred_prob_GB)
roc_auc_GB = roc_auc_score(y_clas_test_bin, y_clas_pred_prob_GB)
plt.plot(fpr_GB, tpr_GB, label='Gradient Boosting (AUC = %0.2f)' % roc_auc_GB)

# AdaBoost Classifier
y_clas_pred_prob_ABC = ABC.predict_proba(X_clas_test)[:, 1]
fpr_ABC, tpr_ABC, thresholds_ABC = roc_curve(y_clas_test_bin, y_clas_pred_prob_ABC)
roc_auc_ABC = roc_auc_score(y_clas_test_bin, y_clas_pred_prob_ABC)
plt.plot(fpr_ABC, tpr_ABC, label='AdaBoost (AUC = %0.2f)' % roc_auc_ABC)

# Bagging Classifier
y_clas_pred_prob_BC = BC.predict_proba(X_clas_test)[:, 1]
fpr_BC, tpr_BC, thresholds_BC = roc_curve(y_clas_test_bin, y_clas_pred_prob_BC)
roc_auc_BC = roc_auc_score(y_clas_test_bin, y_clas_pred_prob_BC)
plt.plot(fpr_BC, tpr_BC, label='Bagging (AUC = %0.2f)' % roc_auc_BC)

# Extra Trees Classifier
y_clas_pred_prob_ET = ET.predict_proba(X_clas_test)[:, 1]
fpr_ET, tpr_ET, thresholds_ET = roc_curve(y_clas_test_bin, y_clas_pred_prob_ET)
roc_auc_ET = roc_auc_score(y_clas_test_bin, y_clas_pred_prob_ET)
plt.plot(fpr_ET, tpr_ET, label='Extra Trees (AUC = %0.2f)' % roc_auc_ET)

# Plotting ROC curve
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc='lower right')
plt.show()

# AUC Scores
print('AUC Scores:')
print('Random Forest:', roc_auc_RFC)
print('Logistic Regression:', roc_auc_LRC)
print('SVM:', roc_auc_SVM)
print('Gradient Boosting:', roc_auc_GB)
print('AdaBoost:', roc_auc_ABC)
print('Bagging:', roc_auc_BC)
print('Extra Trees:', roc_auc_ET)

```
