```python
#import liabraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

```python
#finding the best random state

MaxAccu = 0
MaxRs = 0

for i in range(1, 200):
    X_reg_train, X_reg_test, y_reg_train, y_reg_test = train_test_split(X_reg, y_reg, test_size=0.3, random_state=i)
    lr = LinearRegression()
    lr.fit(X_reg_train, y_reg_train)
    pred = lr.predict(X_reg_test)
    acc = r2_score(y_reg_test, pred)
    if acc > MaxAccu:
        MaxAccu = acc
        MaxRs = i

print("Maximum R2 score is", MaxAccu, "on Random State", MaxRs)
```

```python
#Separating features and label for classification
X_clas = df.drop("primary_fuel",axis=1)
y_clas= df["primary_fuel"]
```

```python
#Feature Scalling using standard scalarization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_clas = pd.DataFrame(scaler.fit_transform(X_clas), columns=X_clas.columns)
X_clas
```

```python
#checking variance inflation factor (vif)
from statsmodels.stats.outliers_influence import variance_inflation_factor

Vif = pd.DataFrame()
Vif['VIF Values'] = [variance_inflation_factor(X_clas.values, i) for i in range(len(X_clas.columns))]
Vif['Features'] = X_clas.columns
Vif
```

```python
#value count in target variable
y_clas.value_counts()
```

```python
#Oversampling the data
from imblearn.over_sampling import SMOTE

SM = SMOTE()
X_clas, y_clas= SM.fit_resample(X_clas, y_clas)
```

```python
y_clas.value_counts()
```

```python
#import liabraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
#Finding the best randam state

maxAccu = 0
maxRs = 0

for i in range(1, 200):
    X_clas_train, X_clas_test, y_clas_train, y_clas_test = train_test_split(X_clas, y_clas, test_size=0.3, random_state=i)
    RFC = RandomForestClassifier()
    RFC.fit(X_clas_train, y_clas_train)
    pred = RFC.predict(X_clas_test)
    acc = accuracy_score(y_clas_test, pred)
    if acc > maxAccu:
        maxAccu = acc
        maxRs = i

print("Best accuracy is", maxAccu, "at random_state", maxRs)
```

```python
X_clas_train, X_clas_test, y_clas_train, y_clas_test = train_test_split(X_clas,y_clas, test_size=0.3, random_state=maxRs)
```

```python
# Import libraries
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, GradientBoostingClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, accuracy_score
```
